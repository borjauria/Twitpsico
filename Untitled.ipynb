```python
from tweepy.streaming import StreamListener
from tweepy import OAuthHandler
from tweepy import Stream
import tweepy
import datetime
import time
```

### Twitter API credentials

```python
consumerKey = ""
consumerSecret = ""
accessToken = ""
accessTokenSecret = ""
```

```python
# creating authentication 
auth = tweepy.OAuthHandler(consumerKey, consumerSecret)
auth.set_access_token(accessToken, accessTokenSecret)

# connecting to the Twitter API using the above authentication
api = tweepy.API(auth)

# definying the data interval
startDate = datetime.datetime(2019, 4, 23, 0, 0, 0)
endDate =   datetime.datetime(2019, 4, 27, 0, 0, 0)
```

### Function that allows Stream tweets and save them in CSV files

```python
class listener(StreamListener):

    def on_data(self, data):
        try:
            tweet = data.split(',"text":"')[1].split('","source')[0]
            print(tweet)
            
            save_date= str(time.time())+'::'+tweet
            #save data into csv file
            save_file = open('tweets_import.csv', 'a')
            save_file.write(save_date)
            save_file.write('\n')
            save_file.close()
            return(True)
        
        except BaseException:
            print('Failed on data')
            time.sleep(5)
            
    def on_error(self, status):
        print(status)
```

### Getting NEGATIVE tweets

```python
twitterStream = Stream(auth, listener())
twitterStream.filter(languages=['es'], track=['madrid'])
```
